```python
#line chart
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [7,8,9,10,11]
plt.plot(x,y,color='magenta',linewidth="5",label ="line") #linestyle for dashed dotted etc.
plt.title("Demo of Line chart")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.axis([0,11,0,11])
plt.show()
```

```python
#bargraph
x = [1,2,3,4,5,6,7]
y = [10,11,12,13,14,15,16]
plt.bar(x,y,color='cyan',width=1,label='bar',edgecolor='black')
plt.axis([0,10,0,20])
plt.title("Demo of bar graph")
plt.xlabel = ("x")
plt.ylabel = ("y")
plt.legend()
plt.show()
```

```python
#scatterplot
x = [1,2,3,4,5]
y = [1,2,1,2,1]
plt.plot(x,y)
plt.scatter(x,y,color='green',label="scatter",marker="*",s=100)
plt.title("Demo of scatter plot")
plt.xlabel=("x")
plt.ylabel=("y")

plt.show()
```

```python
#histogram
x = [22, 87, 5, 43, 56,
              73, 55, 54, 11,
              20, 51, 5, 79, 31,
              27]
bins = [0, 25, 50, 75, 100]
plt.hist(x,bins)
plt.show()
```
